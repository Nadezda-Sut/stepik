```python
import string

n = 'once upon a time, in a land far far away lived a princess , whose beauty was yet unmatched'
m = []
b = ''
counter = -1
while len(n) != 0:
  if n[0] in string.ascii_lowercase:
    b += n[0]
    n = n[1:]
    continue
  if n[0] == ',':
    if len(b) == 0:
      m[-1] = m[-1][:-1] + ', '
      n = n[1:]
    else:
      if len(b) > counter:
        counter = len(b)
      b += ', '
      n = n[1:]
      m.append(b)
      b = ''
      continue
  elif n[0] == ' ':
    if len(b) == 0:
      n = n[1:]
    else:
      if len(b) > counter:
        counter = len(b)
      n = n[1:]
      m.append(b + ' ')
      b = ''

if b:
    m.append(b)
    if len(b) > counter:
        counter = len(b)

l = counter * 3


a = []
f = ''
for i in m:
  if ' ' in i and len(i) == l - len(f) + 1:
    f = f + i[:-1]
    a.append(f)
    f = ''
  elif len(i) <= l - len(f):
    f = f + i
    if len(f) == l:
      a.append(f.strip())
      f = ''
  elif len(i) > l - len(f):
    a.append(f.strip())
    f = i
if f:
  a.append(f.strip())

print(*a, sep = '\n')
```
